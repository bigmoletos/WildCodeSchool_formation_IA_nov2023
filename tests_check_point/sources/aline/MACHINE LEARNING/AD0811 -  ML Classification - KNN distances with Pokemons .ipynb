```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
```

```python
poke = pd.read_csv('https://raw.githubusercontent.com/murpi/wilddata/master/pokemon.csv', index_col=0)
poke.head()
```

```python
# dataframe X avec les valeurs numériques des pokémon non légendaires
poke_sans_legendaires = poke[poke['Legendary'] == False]
X = poke_sans_legendaires[['HP', 	'Attack', 	'Defense',	'Sp. Atk', 	'Sp. Def', 	'Speed']]
X
```

```python
#entrainement du modèle
distanceKNN = NearestNeighbors(n_neighbors=2).fit(X)
```

```python
#affichage des paramètres des pokemon à remplacer
mes_légendaires = poke[(poke['Name'] == 'Mewtwo') | (poke['Name'] == 'Lugia') | (poke['Name'] == 'Rayquaza') | (poke['Name'].str.contains('Giratina')) | (poke['Name'] == 'Dialga') | (poke['Name'] ==  'Palkia')]
mes_légendaires
```

```python
#enregistrement des distances et voisions les plus proches dans des variables
neighMewtwo = distanceKNN.kneighbors([[106, 	110, 	90, 	154, 	90, 	130]])
neighLugia = distanceKNN.kneighbors([[106, 	90, 	130, 	90, 	154, 	110]]	)
neighRay = distanceKNN.kneighbors([[105, 	150 ,	90, 	150 ,	90 ,	95]])
neighDialga = distanceKNN.kneighbors([[100, 	120, 	120 ,	150 ,	100 ,	90]])
neighPalkia = distanceKNN.kneighbors([[90, 	120 ,	100 ,	150 ,	120 ,	100]])
neighGiratinaAlt = distanceKNN.kneighbors([[150, 	100 ,	120 ,	100 ,	120 ,	90]])
neighGiratinaOr = distanceKNN.kneighbors([[150, 	120 ,	100 ,	120 ,	100 ,	90 ]])
```

```python
#création d'un nouveau dataframe avec les deux voisins les plus proches pour chaque pokémon, priorisés par les distances
recommendation = pd.DataFrame([ ['Mewtwo',   poke_sans_legendaires.iloc[neighMewtwo[1][0][0]]['Name'],  poke_sans_legendaires.iloc[neighMewtwo[1][0][1]]['Name']], 
                                ['Lugia',    poke_sans_legendaires.iloc[neighLugia[1][0][0]]['Name'],   poke_sans_legendaires.iloc[neighLugia[1][0][1]]['Name']],
                                ['Rayquaza', poke_sans_legendaires.iloc[neighRay[1][0][0]]['Name'],     poke_sans_legendaires.iloc[neighRay[1][0][1]]['Name']],
                                ['Dialga',   poke_sans_legendaires.iloc[neighDialga[1][0][0]]['Name'],  poke_sans_legendaires.iloc[neighDialga[1][0][1]]['Name']],
                                ['Palkia',   poke_sans_legendaires.iloc[neighPalkia[1][0][0]]['Name'],  poke_sans_legendaires.iloc[neighPalkia[1][0][1]]['Name']],
                                ['Giratina Altered Forme',   poke_sans_legendaires.iloc[neighGiratinaAlt[1][0][0]]['Name'],  poke_sans_legendaires.iloc[neighGiratinaAlt[1][0][1]]['Name']],
                                ['Giratina Origin Forme',   poke_sans_legendaires.iloc[neighGiratinaOr[1][0][0]]['Name'],  poke_sans_legendaires.iloc[neighGiratinaOr[1][0][1]]['Name']]
                               ],
                               columns = ['pokemon_à_remplacer', 'first_Choice', 'second_Choice']
                              )
recommendation
```
