# Mission 4

Find 3 ways to return the following string without including the spaces at the beginning and the end: " Data Analyst ".

```
#first way : using strip() function
string = " Data Analyst "
print(string.strip())
```

```
#second way : using Regex
import re
print(re.sub(r"^\s+|\s+$", "", string), sep='')
```

```
#third way : using substring
string = " Data Analyst "
newString = string[1:-1]
print(newString)
```

# Mission 5

Create a script that converts all vowels to uppercase and all consonants to lowercase.

```
print("enter a phrase")
string = input()

convertString = ""
for letter in string:
  if letter in "aeiouAEIOU":
    convertString += letter.upper()
  else:
    convertString += letter.lower()
print(convertString)
```
