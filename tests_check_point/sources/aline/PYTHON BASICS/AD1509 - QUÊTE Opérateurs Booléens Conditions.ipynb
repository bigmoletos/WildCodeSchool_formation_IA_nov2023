```python
borne1 = int(input("Saisissez un premier nombre entier"))
borne2 = int(input("Saisissez un deuxième nombre entier"))
print("Merci l'intervalle est compris entre", borne1, "et", borne2)
entrée = int(input("Entrez maintenant un troisième nombre entier"))

#premier cas de figure : si la borne1 est plus petit que la borne2
if borne1 < borne2:
  if entrée < borne1:
    print("Tu as entré", entrée, "plus petit que", borne1,"la borne inférieure")
    print("Résultat:", borne1)
  elif entrée > borne2:
    print("Tu as entré", entrée, "plus grand que", borne2, "la borne supérieure")
    print("Résultat:", borne2)
  else:
    print("Tu as entré", entrée, "qui est dans l'intervalle")
    print("Résultat:", entrée)

#deuxième cas de figure : si la borne1 est plus grande ou égale à la borne2
else:
  if entrée < borne2:
    print("Tu as entré", entrée, "plus petit que", borne2,"la borne inférieure")
    print("Résultat:", borne2)
  elif entrée > borne1:
    print("Tu as entré", entrée, "plus grand que", borne1, "la borne supérieure")
    print("Résultat:", borne1)
  else:
    print("Tu as entré", entrée, "qui est dans l'intervalle")
    print("Résultat:", entrée)
```
