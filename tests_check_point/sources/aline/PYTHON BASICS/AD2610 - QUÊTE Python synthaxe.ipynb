 Une fonction qui convertit des montants en dollars vers des montants en euros (prends le cours euro/dollar du jour où tu fais la quête).

```python
def dollar_to_euro(num):
  return num*0.86
print(dollar_to_euro(3))

print((lambda num: num*0.86)(3))
```

Une fonction qui renvoie la somme des nombres pairs présents dans une liste contenant des nombres entiers.

```python
def sumClassique(liste):
  somme =0
  for i in liste:
    if i%2==0:
      somme += i
  return somme
print(sumClassique([1,2,3,4]))


(lambda liste : sum(i for i in liste if i%2 == 0))([1,2,3,4])
```

Un script qui convertit une liste de nombres entiers positifs ou négatifs) en une nouvelle liste contenant les valeurs absolues.

```python
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]

def valeurAbs(liste):
  newListe = []
  for i in liste:
    if i<0:
      newListe.append(-i)
    else :
      newListe.append(i)
  return newListe
print(valeurAbs(test_list))

(lambda liste : [-i if i<0 else i for i in liste])(test_list)
```
