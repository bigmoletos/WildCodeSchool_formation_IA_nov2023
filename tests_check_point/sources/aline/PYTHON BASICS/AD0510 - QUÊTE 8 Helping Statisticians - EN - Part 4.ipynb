```
#power of x
def myXPowerFunction(a, b):
  if b==0:
    result = 1
  result = 1
  for power in range(b):
    result = result*a
  return result

myXPowerFunction(10,5)
```

```
#prime number
def prime(x):
  modulo = []
  primenumber = []
  for num in range(0,x):
    if num == 0 or num == 1:
      continue
    modulo = []
    for diviseur in range(2,num):
      modulo.append(num%diviseur)
    if 0 not in modulo:
      primenumber.append(num)
      
  return primenumber

prime(20)
```

```
#Continuation of Fibonnaci
def fibonnaci():
  liste = [0,1]
  for n in range(2,30):
    f= liste[-1] + liste[-2]
    liste.append(f)
  return liste

fibonnaci()
```

# The central theorem limits

The code below simulates the **central limit theorem**. You must replace the following functions with your own functions all the while achieving the same visual result :


* sqrt(): Your square root function
* pow() : Your power function of x



```
# Code example

import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

domain = range(-100,100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domain]
plot = plt.plot(domain, y)
```
