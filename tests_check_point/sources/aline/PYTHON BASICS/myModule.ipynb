```
#Square of a number:
def square(n):
  return n**2

square(4)
```

```
#Cube of a number
def cube(n):
  return n**3

cube(5)
```

```
#Absolute value
def abs(n):
  if n >= 0:
    return n
  else :
    return -n

abs(-34)
```

```
#Factorial of a number
def factorial(n):
  f = 1
  for i in range(1,n+1):
    f = f*i
  return f

factorial(4)
```

```
#Mode of a list of numbers
def mode(liste):
  dico = {}
  for i in liste :
    dico[i]=liste.count(i)
  return max(dico, key=dico.get)

mode([10, 20, 20, 30, 40, 40, 40, 10, 30, 20, 20])
```

```
#Average of a list of numbers
def average(liste):
  return sum(liste)/len(liste)
  
average([10, 20, 20, 30, 40, 40, 40, 10, 30, 20, 20])
```

```
def minimum(liste):
  minimum = liste[0]
  for i in liste:
    if i < minimum:
      minimum = i
  return minimum

minimum([10, 20, 20, 30, 40, 40, 40, 10, 30, 20, 20])
```

```
def maximum(liste):
  maximum = liste[0]
  for i in liste:
    if i > maximum:
      maximum = i
  return maximum

maximum([10, 20, 20, 30, 40, 40, 40, 10, 30, 20, 20])
```
