```python
from random import choice, randint
```

```python
class dataAnalyst:
  def __init__(self, nom, prenom, age, sexe, formationPrecedente):
    self.nom = nom
    self.prenom = prenom
    self.age = age
    self.sexe = sexe
    self.formationPrecedente = formationPrecedente
    self.motivation = 100
    self.progression = 0


  def suivreFormation(self):
      self.motivation -= randint(5,25)
      self.progression += randint(20,30)
      return
    
  def bosserPlus(self):
      self.motivation += randint(10,30)
      self.progression += randint(10,30)
      return

  def echouer(self):
      self.motivation -= randint(20,40)
      self.progression += randint(20,40)
      return

  def reussir(self):
      self.motivation += randint(20,40)
      self.progression += randint(10,20)
      return

  def studyRoutine(self) -> str:
    listeMethode = [self.suivreFormation, self.bosserPlus, self.echouer, self.reussir]
    nomDeLaMethode = choice(listeMethode)
    nomDeLaMethode()
    return nomDeLaMethode.__name__ 
```

```python

aline = dataAnalyst('DUBUY','Aline',28, 'F', 'entrepreneuriat social')
```

```python
tour = 0
while aline.motivation > 0 and aline.progression < 100:
  tour += 1
  methode = aline.studyRoutine()
  print(f'Tour {tour}')
  print(f'La méthode {methode} a été utilisée.')
  print(f'Motivation vaut maintenant {aline.motivation}')
  print(f'Progression vaut maintenant {aline.progression}')
  print('')
if aline.motivation < 0:
  print("BRAVO TU AS GAGNÉ !!!")
else : 
  print("BRAVO TU AS APPRIS !!!")
```
