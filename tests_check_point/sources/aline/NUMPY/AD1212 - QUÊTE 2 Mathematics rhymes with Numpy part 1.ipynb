## Mission 3

Create a function that returns the [inverse value (for `3`, return `1/3`)](https://en.wikipedia.org/wiki/Multiplicative_inverse) of a set of values contained in a NumPy array

```
def inverse(arr, set_start,set_stop):
  for value in arr:
    index = np.where(arr == value)
    arr[index] = 1/value
  return arr[set_start : set_stop]

arrTest = np.arange(1,10, dtype=float)
inverse(arrTest, 5, 9)
```
