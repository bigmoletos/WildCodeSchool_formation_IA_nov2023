# Matplotlib

2 différentes interfaces : 
* Interface Matlab
* Interface Orientée objet

```python
# une seule instance plt de pyplot pour tous les affichages
x = [1, 2, 3, 4]
y = [3, 6, 8, 12]
plt.plot(x,y)
plt.title('notre premier schema')
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.show()
```

```python
# 2 objets, figure and axes
x = [1, 2, 3, 4]
y = [3, 6, 8, 12]
fig, ax = plt.subplots(figsize=(16, 8), nrows=1, ncols=2)
print(ax)
ax[0].plot(x,y)
ax[0].set_xlabel("1er plot")
ax[1].plot(np.array(x)+1, y)
ax[1].set_yticks([0, 5, 8, 12])
plt.show()
```

```python
dir(ax[0])
```

Recommandation

```python
fig = plt.figure(figsize=(16,8)) # creer une figure
ax1 = fig.add_subplot(211) # definir le nombre de subplot et creer les axes
ax2 = fig.add_subplot(212)

# modifier chaque ax pour faire l'affichage
ax1.set_xlabel('x')
ax1.set_ylabel("y")
ax1.plot(x, y)


ax2.set_xlabel("Category")
ax2.set_ylabel("Taille")
x=['Hommes','Femmes']
y=[180,160]
ax2.bar(x,y)
```

# Guide de visualisation

* une variable numerique : histogramme, boxplot, kernel density plot
* une variable catégorielle: piechart, barchat

* 2 variables numériques : scatter plot, heatmap
* 2 variables catégorielles: barchat

* 1 variable numérique, 1 variable catégorielle : Violin plot
* 1 variable numérique , 2 variables catégorielles : Facet grid



# Seaborn

## Box plot (analyse univarié numérique)

```python
titanic.head()
```

```python
sns.boxplot(data=titanic, x="Age")
```

```python
sns.boxplot(data=titanic, y="Age")
```

```python
sns.boxplot(titanic['Age'])
```

## Kernel density plot (variable numérique)

```python
sns.kdeplot(titanic.Age)
```

## Violin plot (x=category, y=numerique)

```python
sns.violinplot(x="Survived", y="Age", data=titanic)
```

## Countplot (une variable catégorielle)

```python
# tidy data
sns.countplot(x = "Sex", data= titanic)
```

## Heatmap (plusieurs variables numériques)

```python
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu')
```

```python
sns.heatmap(titanic.corr()[['Survived', 'Age']], annot=True, cmap='YlGnBu')
```

## Facetgrid (2 variable categorie et 1 numérique)

```python
cond = (titanic.Sex=='male') & (titanic.Survived==0)
titanic[cond]['Age'].hist()
```

```python
grid = sns.FacetGrid(titanic, col="Sex", row="Survived")
grid.map(plt.hist, 'Age')
```

## Regplot (scatter plot du riche)

```python
sns.regplot(data=titanic, x='Age', y='Fare')
```

## lmplot (facetgrid + regplot) 2 variables category 2 variable numérique

```python
titanic.Fare.hist()
```

```python
sns.lmplot(data=titanic, x='Age', y='Fare', row='Survived', col='Sex')
```

## Strip plot : Scatter mais pour visualiser une variable numérique et une variable de catégorie. Une sorte de boxplot avec des points

```python
sns.stripplot(data=titanic, x="Survived", y='Age')
```

## swarm plot : strip plot mais avec la distribution

```python
sns.swarmplot(data=titanic, x="Survived", y='Age')
```

## Catplot = strip plot + facet grid

```python
titanic.head(2)
```

```python
sns.catplot(data=titanic, x="Survived", y="Age", col="Sex", row="Pclass")
```

```python
sns.catplot(data=titanic, x="Survived", y="Age", col="Sex", row="Pclass", kind='violin')
```

```python
sns.catplot(data=titanic, x="Survived", y="Age", col="Sex", row="Pclass", kind='swarm')
```
