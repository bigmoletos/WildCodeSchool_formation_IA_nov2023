```python
import random


def triangle_plein():
  while True:
    hauteur_triangle = input("Veuillez entre la hauteur de triangle: ")
    if hauteur_triangle.isdigit():
      hauteur_triangle = int(hauteur_triangle)
    else:
      hauteur_triangle = traduire_texte_chiffre(hauteur_triangle)
    insertion_parasite = False
    parasite = random.randint(0, 10)

    # controle de la hauteur
    if hauteur_triangle < 3 or hauteur_triangle > 10:
      print("Erreur : veuillez saisir une hauteur minimale de 3 et maximale de dix ")
    else:
      symbol = input("Veuillez entre le symbole constituant votre triangle: ")
      # impression triangle
      for i in range(1, hauteur_triangle+1):
        message = symbol * i
        if not insertion_parasite and i == random.randint(2, hauteur_triangle-1):
          pos_parasite = random.randint(0, len(message)-1)
          message = message[:pos_parasite] + \
              str(parasite) + message[pos_parasite+1:]
          insertion_parasite = True
        print(message)

    message = input("""Voulez-vous arreter le programme ou continuer ?
                  tapez Oui(o) pour arreter
                  tapez sur n'importe quelle touche pour continuer:  """)

    if message.lower() in ['o', 'oui', 'y', 'yes']:
        break


def traduire_texte_chiffre(texte):
  dico = {
      'un': 1,
      'deux': 2,
      'trois': 3,
      'quatre': 4,
      'cinq': 5,
      'six': 6,
      'sept': 7,
      'huit': 8,
      'neuf': 9,
      'dix': 10
  }
  chiffre = dico.get(texte.lower(), 0)
  return chiffre


# #  test
# triangle_plein()
```

Créez une **fonction** qui dessinera un triangle rectangle plein (voir ci-dessous) remplie d'une valeur choisie par l'utilisateur. Nous pouvons indiquer la hauteur du triangle (3 étant le minimum). Tant qu'il n'indique pas une hauteur valide du triangle (soit 3, ou alors une valeur valide) le programme recommance sans qu'il ne soit nécessaire de relancer la cellule.

Le programme doit placer aléatoirement un nombre entre 0 et 9 à l’intérieur du triangle mais pas sur ses extrémités (ce qui exclu les deux 1er niveau, d'où une hauteur minimale de 3).

Le programme vérifiera que l'utilisateur a bien renseigné un nombre pour la hauteur du triangle. Si il renseigne un nombre dans une chaine de caractère il faudra la transtyper. Si il renseigne autre chose il faudra lui redemander une valeur valide sans qu'il ne soit nécessaire de relancer la celulle.

Il peut écrire en texte les chiffre (zéro, un, deux...dix). Dans ce cas la valeur sera remplacée par le int correspondant. Donc si il indique pour la hauteur > "neuf", "Neuf" ou "NEUF" il faudra générer un triangle d'une hauteur de 9.

Enfin vous n'utiliserez que des boucles while, pas de boucle for.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def monty_hall(nombre_parties):

  # nombre de parties
  # nombre_parties=100000
  # nombre_parties=np.random.randint(10,1000,100)*100

  # utilisation d'un generator python
  def alternate_strategy():
      while True:
          yield "change"
          yield "keep"

  def play_games(strategy, nb_turns):
      # Initialiser les portes et les choix
      correct_doors = np.random.randint(0, 3, nb_turns) #choix 0 1 2 possibles
      first_choices = np.random.randint(0, 3, nb_turns)
      strategy_generator = alternate_strategy() # execution de la fonction de generation de keep ou change alernativement
      second_choices_success = np.zeros(nb_turns, dtype=bool)

      for i in range(nb_turns):
          # Si la stratégie est de changer, le deuxième choix est réussi si le premier choix était incorrect
          # Si la stratégie est de garder, le deuxième choix est réussi si le premier choix était correct
          if strategy == "alternate":
              current_strategy = next(strategy_generator)   # si nous avions change alors il mettra keep
              if current_strategy == "change":
                  second_choices_success[i] = (first_choices[i] != correct_doors[i])
              elif current_strategy == "keep":
                  second_choices_success[i] = (first_choices[i] == correct_doors[i])
          elif strategy == "change":
              second_choices_success[i] = (first_choices[i] != correct_doors[i])
          elif strategy == "keep":
              second_choices_success[i] = (first_choices[i] == correct_doors[i])
          else:
              raise ValueError("Strategy not recognized!")

      return second_choices_success

# Jouer plusieurs parties avec différentes stratégies
  play_change = np.sum(play_games("change", nombre_parties))
  play_keep = np.sum(play_games("keep", nombre_parties))

 # Pour la stratégie "alternate", nous générons un tableau de stratégies aléatoires ("change" ou "keep") pour chaque partie
  play_alternate = np.sum(play_games("alternate", nombre_parties))


  plt.figure(figsize=(10, 5))

  # Créer le graphique à barres
  plot = plt.bar([1, 2, 3],
                [play_change, play_keep, play_alternate],
                tick_label=["Change", "Garde", "Alternate"],
                color=['red', 'yellow', 'green'])

  # Ajouter un titre au graphique avec le nombre de parties
  plt.title(f"Résultats des différentes stratégies avec {nombre_parties:,} parties".replace(',', ' '))

  # Ajouter des titres aux axes
  plt.xlabel("Stratégie")
  plt.ylabel("Nombre de victoires ")

  # Afficher le graphique
  plt.show()
```
