```
# 1 Puissance de x : Créez une fonction qui prend deux nombres comme paramètres, un nombre et une puissance, et ramène ce nombre à la puissance indiquée.

# 1 Square of a number: Create a function that returns the square of a number.
def myXPowerFunction(nombre, puissance):
  resultat=1
  for _ in range(puissance):
    resultat*=nombre
  return resultat

#-----------
# test de la fonction
nombre_utilisateur=2
puissance_utilisateur=8
print(myXPowerFunction(nombre_utilisateur,puissance_utilisateur))


```

```
# 2 Nombre premier inférieur à x : Créez une fonction qui prend un nombre et renvoie tous les nombres premiers qui lui sont inférieurs.
import random
def get_nombres_premiers_inferieurs(nombre):
  if nombre < 2:
    return False
  racine_carree_nombre=nombre ** 0.5  # calcul de la racine carrée de nombre
  resultats=[]  #initialiser le tableau des résultats
  # resultats=[None]*int(racine_carree_nombre) # crée un tableau de n elements vide
  # resultats=list(range(int(racine_carree_nombre))) # crée un tableau de n elements
  #création d'une liste de chiffres premiers avec le postulat que
  # si le nombre est inférieur à 2 il n'a pas de chiffre premier et
  # si la racine carré de n est supérieure au nombre alors on a un chiffre premier, on peut aussi pour faire la racine carrée on fait la puissance 0.5
  liste_chiffre_premier= [nombre_premier for nombre_premier in range(2, nombre + 1) if all(nombre_premier % i != 0 for i in range(2, nombre_premier) if i * i <= nombre_premier)] # créer une liste en compréhension des nombres premier
  # print("liste",liste_chiffre_premier)

  for i in range(2,nombre):
    if i in liste_chiffre_premier:
      resultats.append(i)

  return resultats

#-----------
# test de la fonction
nombre_utilisateur=random.randint(0,100)

print("la liste des chiffres premiers inférieur à :",nombre_utilisateur," sont :",get_nombres_premiers_inferieurs(nombre_utilisateur))

```

```
# 3 Suite de Fibonnaci : Créez une fonction qui ne prend aucun paramètre, mais renvoie les 30 premiers nombres de la séquence de Fibonnaci.
import random
def suite_fibonnaci(nombre):
  # if nombre<0:
  #   return
  a,b=0,1
  resultat=[]
  for i in range(1, abs(nombre+1)):
    c=a+b
    resultat.append(b)
    a=b
    b=c
  # resultat=b

  return resultat

#-----------
# test de la fonction
nombre_utilisateur=random.randint(-10,10)

print("la suite de fibonnaci jusqu'à :",nombre_utilisateur," est :",suite_fibonnaci(nombre_utilisateur))

```

```
# 4 Racine carrée : Créez une fonction qui renvoie la racine carrée d'un nombre.
import random
def myFunctionRacineCarre(nombre):
  if nombre < 0:
    return None  # ici on a None c'est pour que notre code de matplot fonctionne
  racine_carree_nombre=nombre ** 0.5  # calcul de la racine carrée de nombre
  resultat=round(racine_carree_nombre,2)

  return resultat

#-----------
# test de la fonction
nombre_utilisateur=random.randint(-10,20)

print("la racine carrée de :",nombre_utilisateur," est :",myFunctionRacineCarre(nombre_utilisateur))

```
