# Tasks

Normalize pixel values of both train and test images to the values in the range between 0 and 1

```python
# your code
#  chaque pixel d’une image est représenté par une valeur d’intensité entre 0 et 255 (pour une image en niveaux de gris).
#  La normalisation des images consiste à diviser chaque pixel par 255, ce qui transforme la plage de chaque pixel de [0,255] à [0,1]
train_images, test_images = train_images / 255.0, test_images / 255.0
```

Build a sequential model with the following architecture:

conv2d - (None, 30, 30, 32)  
_________________________________________________________________
max_pooling2d - (None, 15, 15, 32)        
_________________________________________________________________
conv2d - (None, 13, 13, 64)
_________________________________________________________________
max_pooling2d - (None, 6, 6, 64)        
_________________________________________________________________
conv2d - (None, 4, 4, 128)
_________________________________________________________________
flatten - (None, 2048)       
_________________________________________________________________
dense - (None, 64)    
_________________________________________________________________
dense - (None, 10)       


```python
# your code
# Création du modèle
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Construction du modèle
reseau = Sequential()
# sortie : (None, 30, 30, 32)
reseau.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
reseau.add(MaxPooling2D((2, 2)))  # sortie : (None, 15, 15, 32)
# sortie : (None, 13, 13, 64)
reseau.add(Conv2D(64, (3, 3), activation='relu'))
reseau.add(MaxPooling2D((2, 2)))  # sortie : (None, 6, 6, 64)
# sortie : (None, 4, 4, 128)
reseau.add(Conv2D(128, (3, 3), activation='relu'))
reseau.add(Flatten())  # sortie : (None, 2048)
reseau.add(Dense(64, activation='relu'))  # sortie : (None, 64)
reseau.add(Dense(10))  # sortie : (None, 10)
reseau.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

Compile the model using Adam optimizer, sparse categorical crossentropy as loss function, and choose an appropriate metric for classification. Use 10 epochs to train the model.

```python
# your code
history = reseau.fit(train_images, train_labels, epochs=10)
```

Evaluate the model on test data to get the loss and accuracy metrics

```python
# your code
# Suppose 'X_test' and 'y_test' are your test data and labels
test_loss, test_accuracy = reseau.evaluate(test_images, test_labels)

print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```
