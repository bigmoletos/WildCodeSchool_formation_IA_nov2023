```python
# Mission 1: use apply to create a new column called 'reviews_opinion': if the ratings is between 4.5 and 5, the reviews_opinion is 'very good';
# if the rating is between 4 and 4.5, the reviews_opinion is 'good'; if the rating is between 3 and 4, the reviews_opinions will be 'neutral';
# if the rating is up to 3 the reviews_opinion is going to be 'bad'

# Add your code here

# fonction de regroupement des notes
def range_rating(average_rating):
  if average_rating <= 3:
    return 'bad'
  if average_rating <= 4 :
    return 'neutral'
  if average_rating <= 4.5:
    return 'good'
  if average_rating <= 5:
    return 'very good'
  else:
    return 'excellent'
```

```python
# test fonction
aver = 4.6
print(range_rating(aver))
```

```python
# ajout de la nouvelle colonne depuis la fonction avec apply
# par sécu nous faisons une copie du dataframe
books2=books.copy()
books2['reviews_opinion'] = books2['average_rating'].apply(range_rating)
books2.head(5)
```

```python
# Mission 2: using apply with a function, create a column with the number of authors per book
# (note that in the column 'authors', each name of the author is divided with the character '/')

# Add your code here
def number_of_authors_per_books(names):
    return len(names.split('/'))


books21 = books2.copy()

books21['num_authors'] = books21['authors'].apply(
    number_of_authors_per_books )
books21.head(5)
```

```python
# Mission 3: rewrite the same task as Mission 2 using the lamdba function
# deja fait avec une lambda
# Add your code here
books2['num_authors'] = books2['authors'].apply(lambda x: len(x.split('/')))
books2.head(5)
```

```python
# Mission 4: create a column 'comment_ration'.
# Use the columns 'ratings_count' and 'text_reviews_count' to calculate the percentage of how many text review per reviews are per book:
# percentage = text / reviews * 100
# Use the mathematical operations between columns

# Add your code here
books3=books2.copy()
books3['comment_ration'] = (
    books3['text_reviews_count']/books3['ratings_count'])*100

books3.head(5)
```

```python
# Mission 5: obtain the same result of Mission 4 using apply with a function.
# Try both methods to apply to the whole dataset and apply to selected columns

# Add your code here
def ratio_of_review(row):
    return (row['text_reviews_count'] / row['ratings_count'])*100

books4=books2.copy()
books4['comment_ration'] = books4.apply(ratio_of_review, axis=1)

books4.head(5)
```

```python
# Mission 6: obtain the same result of Mission 5 using apply lambda
# Try both methods to apply to the whole dataset and apply to selected columns

# Add your code here

books5 = books2.copy()
books5['comment_ration'] = books5[['text_reviews_count', 'ratings_count']].apply(
    lambda row: (row['text_reviews_count'] / row['ratings_count']) * 100, axis=1)

books5.head(5)
```
