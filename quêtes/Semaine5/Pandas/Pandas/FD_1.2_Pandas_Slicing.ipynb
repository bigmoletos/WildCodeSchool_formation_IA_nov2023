# Mission 5

Rename the columns [of this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv), try using the following 2 different methods to achieve this:
- Setting column names directly when reading the file
- Editing after creating the dataframe ([This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html))

```
names_column=["Ville","Couleur","Forme","Status","Heure"]
df=pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv", names=names_column, header=None,  skiprows=1,encoding='ISO-8859-1')
df.head(5)
```

```
# Autre Methode
# names_column=["Ville","Couleur","Forme","Status","Heure"]

# rÃ©cupÃ©ration des noms de colonnes
names_column=df.columns.tolist()
print(names_column)
# CrÃ©ation de la nouvelle liste de colonnes
names_column2=["Ville2","Couleur2","Forme2","Status2","Heure2"]
# Mapping entre les 2 listes
mapping_names=dict(zip(names_column, names_column2))

# renommage des colonnes
df.rename(columns=mapping_names, inplace=True)
df.head(5)
```
