```python
import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
film = pd.read_csv(link)
```

### INFOS SUR LE DATASET

```python
# INFOS SUR LE DATASET
# Have a look at the dataset first
from pandas.core.arrays.categorical import factorize_from_iterable
df = film
df.info()
df.describe()
print("taille: ", df.size, "valeurs.", " \nDimensions:", df.shape)
print("nb na:", df.isna().all().size, "nb null:", df.isnull().all().size)
print("liste nom des colonnes: \n", df.columns)
print("liste colonnes NA: \n", df[df.isna()].columns)
df[df.isna()].head(5)
```

```python
# print("liste nom des valeurs uniques par ligne: \n", df[df.iloc[:,1:]].value.unique())
listes_colonnes = ['star_rating', 'title',
                   'content_rating', 'genre', 'duration', 'actors_list']
# on retire les colonnes date time et invoice id
listes_colonnes2 = ['star_rating', 'content_rating',
                    'genre', 'duration', 'actors_list']
for col in listes_colonnes2:
    if df[col].dtype == 'object' and col != 'Date' and col != 'Time':

        print(f"Valeurs uniques pour {col} : \n", df[col].unique())
```

## ACTEUR LE PLUS SOUVENT MENTIONNE

```python
import ast
# Convertion de la colonne des acteurs en listes avec la fonction
#  ast.literal_eval converti les chaînes de caractères en listes
# On aurait aussi pu utiliser la fonction split(',')
try:
    df['actors_list'] = df['actors_list'].apply(ast.literal_eval)
except:
    df['actors_list']
# print(df['actors_list'])
# Liste des acteurs avec une liste en compréhension
all_actors = [acteur for acteurs in df['actors_list'].tolist()
              for acteur in acteurs]
# print(all_actors)

# Convertion la liste en une Series pour utiliser la méthode value_counts, cela revient à créer une colonne avec tous les noms d'acteurs
actor_series = pd.Series(all_actors)
# print(actor_series)

# Acteur le plus souvent mentionné dans la liste  actor_series
# value_counts, compte le nombre d’occurrences de chaque valeur.
# idxmax est utilisé pour trouve le max de value_count
most_common_actor = actor_series.value_counts().idxmax()

print("L'acteur le plus souvent mentionné est : ", most_common_actor)
```
