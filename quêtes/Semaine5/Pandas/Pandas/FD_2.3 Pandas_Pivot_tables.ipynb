# PIVOT TABLE

There are 2 syntaxes, choose your preferred :

- As a function : pd.crosstab()

```python
df_weather.describe()
```

```python
# Example
df_weather.pivot_table(values='MAX_TEMPERATURE_C',
                       index='MONTH', aggfunc='mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```python
# Hop !
df_weather.pivot_table(values='MAX_TEMPERATURE_C',
                       index=['MONTH', 'OPINION'], aggfunc='max')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```python
# Hop !
df_weather.head(1)
# formatage de la colonne date
df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
# création des colonnes jour et mois
df_weather['day'] = df_weather['DATE'].dt.day
df_weather['month'] = df_weather['DATE'].dt.month
df_weather.head(10)
df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM',
                       index='day', columns='month')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.** (optionnal)

```python
# Hop !
df_weather.head(1)

# df_weather.head(10)
df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM',
                       index='month', aggfunc='sum')
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```python
# Hop !
df_weather.head(1)
df_weather.pivot_table(values=["MAX_TEMPERATURE_C", "SUNHOUR"],
                       index="month", aggfunc="mean")
```

```python
# option perso
df = df_weather.pivot_table(values=["MAX_TEMPERATURE_C", "SUNHOUR"],
                            index="month", aggfunc={"MAX_TEMPERATURE_C": "max", "SUNHOUR": "median"})
```
