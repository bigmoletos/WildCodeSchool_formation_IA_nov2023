Écrire un programme permettant de calculer la somme des entiers entre 0 et 1000 (tous deux inclus) qui sont divisibles par 2 ou par 5, mais pas par 10 ?

# Mission 3

Écrire un programme permettant de calculer la factorielle d'un nombre choisi par l'utilisateur.

Par exemple, la factorielle de 4 est égale à 4 * 3 * 2 * 1

````python
import math
nombre=int(input("entrez un nombre entier positif: "))
factorielle_nombre=1
for i in range(1,nombre+1):
  factorielle_nombre*=i
  print(i)

print(f"la factorielle de {nombre} est : {factorielle_nombre}")

#autre methode en utilisant les fonctions math
print(f"la factorielle de {nombre} est : {math.factorial(int(nombre))}")
````

# Mission 5

En utilisant la boucle while, écrire un programme qui permet d'additionner des entiers au carré (en commençant par 1) jusqu'à ce que le total dépasse 300. Imprimez le total final et le dernier nombre à être élevé au carré .

````python
nombre=1
resultat=0
limite=300
while resultat <= limite:
 # on additionne à partir de 1 le chiffre au carre
  resultat = resultat + nombre**2
  nombre+=1
  #print("nombre :", nombre, "res: ", resultat)
print(f"le total final est {resultat} et le dernier chiffre élevé au carré avant de dépasser {limite} est {nombre-1 }  ")
````

# Mission 6

Écrire un programme qui permet d'afficher la table de multiplication pour tous les nombres variant entre 1 et 10 (les deux inclus).
 - Si le nombre est inférieur ou égale à 5, on souhaite afficher les résultats de ses multiplications par 1 jusqu'à 10.
 - Sinon on affiche les résultats de multiplication de ce nombre par 1 à 5

 Le résultat sera affiché comme ceci :

image.png

````python
#table de muliplication de 1 à 10
resultat=0
# boucle de 1 à 10
for table in range (1,11):
    #je place réinitialise le tableau à chaque fois sinon il se rempli de toutes les tables
    table_muliplication=[]
    # print("table multiplication",table_muliplication)
    print(f"la table de multiplication de {table} : ")

    # boucle de 1 à  10 ou de 1 à 5 en fonction de la table
    for multiplicateur in range (1,6 if table>=5 else 11):
        resultat=table*multiplicateur
        # j'ajoute à la liste "table_multiplication" le resultat du calcul
        table_muliplication.append(str(resultat))

    #je retire les [] et les '' du tableau en le convertissant
    # en string avec une jointure avec un espace " "
    table_en_texte=" ".join(table_muliplication)
        # print(f"{table} * {multiplicateur}  = {resultat}")
    print(table_en_texte)
````
