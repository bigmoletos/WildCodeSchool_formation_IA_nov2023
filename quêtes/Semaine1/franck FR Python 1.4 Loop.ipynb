Pour faire saisir une valeur à l'utilisateur, tu dois utiliser la fonction input().
Dans un premier temps, tu peux simplement tester la fonction input() dans un google collab pour voir ce qu'il se passe lorsque tu exécutes la cellule de code, après avoir écrit input() dans la cellule de code.
Fais saisir ensuite un troisième entier à l'utilisateur.
Ton programme doit afficher si la 3ème valeur saisie, est:

dans l'intervalle
En dessous de la borne inférieure
Au dessus de la borne supérieure


```python

# j'ai rajouté une boucle de facon à gerer les erreurs de saisies
while True:
    try:
      #input avec suppressions des eventuels espaces
        borne1=int(input("veuillez entrer le premier chiffre: ").strip())
        borne2=int(input("veuillez entrer le deuxième chiffre: ").strip())

        print(f"vous avez entré les chiffres suivants: \n Borne1={borne1}\n Borne2={borne2}")
        borne3=int(input("veuillez entrer le 3éme chiffre: ").strip())
        if borne1 == borne2:
          print("veuillez saisir des valeurs differentess")
        elif borne1>borne2:
          max=borne1
          min=borne2
        else:
          max=borne2
          min=borne1


        if borne3<min and borne3<max:
            print(f"votre 3éme borne= {borne3} est  inférieure à la borne={min} ")
        elif borne3>=min and borne3<=max:
            print(f"votre 3éme borne= {borne3} est  comprise entre les borne={borne1} et la borne={borne2}")
        else:
            print(f"votre 3éme borne= {borne3} est  superieure à la borne2={max} ")
      # gestion des value erreurs par exemple lors de la saisie de caractéres et non de chiffres entier
    except ValueError:
        print("Erreur : veuillez saisir uniquement des chiffres entier")
        #ajout d'un input pour proposer à l'utilisateur de contineur ou de sortir du programme
        message=input("""Vous avez saisi des erreurs voulez-vous arreter le programme ou continuer ?
                  tapez Oui(o) pour arreter
                  tapez sur n'importe quelle touche pour continuer:  """)
        # sortie du programme et donc de la boucle avec controle de la réponse mise en minuscules
        if message.lower()  in ['o', 'oui','y','yes']:
           break
    message=input("""Voulez-vous arreter le programme ou continuer ?
                      tapez Oui(o) pour arreter
                      tapez sur n'importe quelle touche pour relancer:  """)
    # sortie du programme et donc de la boucle avec controle de la réponse mise en minuscules
    if message.lower()  in ['o', 'oui','y','yes']:
      break
```
