Une fois que c'est fait, vous pouvez faire l'exercice suivant:  
Parmi ses variables, lesquels ont une valeur divisible par 2, donc un reste égale à 0 ?
Pour répondre, vous allez devoir tester chacune des variable, et remplacer "..." par le symbole qui convient.

```python
var_1 = 3
var_2 = 6
var_3 = 7
var_4 = 14
var_5 = 12
#liste des variables
list_nom_variable=[var_1,var_2,var_3,var_4,var_5]

# liste en compréhension avec seulement les variables paires
list_paires = [nom_variable for nom_variable in list_nom_variable if nom_variable % 2 == 0]

# Afficher la liste
print(f"On s'exerce ici avec la division euclidienne. Le reste de {list_paires} divisé par 2 est égale à 0.")

#if nom_variable % 2  == 0:
#  print(f"On s'exerce ici avec la division euclidienne. Le reste de {nom_variable} divisé par 2 est égale à 0.")
```
