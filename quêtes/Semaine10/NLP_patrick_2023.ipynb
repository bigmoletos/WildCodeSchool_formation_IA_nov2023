```python
def Stemming(liste):
  stemming = []
  for element in liste:
    stemming.append(PorterStemmer().stem(element))
  return stemming
```

```python
def Lemmatisation(liste):
  lemmatisation = []
  for element in liste:
    lemmatisation.append(WordNetLemmatizer().lemmatize(element))
  return lemmatisation
```

```python
dfProjet["descriptionStemme"] = dfProjet.descriptionTokenise.apply(lambda ligne : Stemming(ligne))
```

```python
dfProjet["descriptionLemmatise"] = dfProjet.descriptionTokenise.apply(lambda ligne : Lemmatisation(ligne))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = dfCategorieAvecAuMoins50Individus.drop("niveau1", axis=1)
y = dfCategorieAvecAuMoins50Individus.niveau1

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2)
```

```python
knn = KNeighborsClassifier().fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
```
