Script number 1

```python
try:
  list_integers = [1, 12, -45, 6, 27, 0, 24,"a"]

  for i in list_integers:
    try:
      print(1/i)

    except ZeroDivisionError:
        # print("Erreur : on ne peut diviser par Zero nous convertissons le résultat en 0")
          print(0)

    except (Exception, NameError) as e:
          print("Une erreur inattendue s'est produite. Veuillez réessayer.\n",e)
except NameError as e:
      print("Veuillez verifier votre liste, elle doit contenir uniquement des chiffres.\n",e)

```

Script number 2

```python
list_integers = [1, 12, -45, 6, 27, 24]
for i in range(10):
  try:
    print(1/list_integers[i])
  except IndexError:
    i=i-1
```

Script number 3

```python
list_integers = [1, 12, -45, 6, 27, 0, 24]
for i in range(10):
  try:
    print(1/list_integers[i])
  except ZeroDivisionError                         :
    print(0)
  except IndexError:
    i-=1
```
