# Mission 2

Ask the user for a string, then for a character (we assume user input follows the rules). Then, display the positions, **starting with 1**, of all occurrences of the character within the string.

For example, for the word "welcome" and for the character "e", the script will display:
```
position: 2
position: 7
```

````
# pour faciliter la correction je ne fais pas les input('veuillez saisir un mot') et  input('veuillez saisir un caractére')
mot='welcome'
caractere_to_find="e"

positions = [carac+1 for carac in range(len(mot)) if mot[carac] == caractere_to_find]
for i in positions:
    print(i)
````

# Mission 6

Ask the user for a string. Then, display the same string with all vowels converted to uppercase, and all consonants converted to lowercase.  
For example:
* string = `"Papa's got a brand new bag"`

````
# pour faciliter la correction je ne fais pas les input('veuillez saisir un mot')
phrase = "Papa's got a brand new bag"
phrase=phrase.lower()
phrase=phrase.replace("a","A").replace('o','O').replace('e','E')
print(phrase)
````
