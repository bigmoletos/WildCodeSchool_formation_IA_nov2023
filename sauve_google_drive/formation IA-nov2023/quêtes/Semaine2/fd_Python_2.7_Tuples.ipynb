# Mission 4

Create 2 tuples of the same length that contain only integers.

Then create a script that compares the sum of their elements and displays the tuple that has the highet total.

```
import random
# creation d'un liste aléatoire de 8 integers
tuple1 = [random.randint(0, 50) for v in range(8)]
print(tuple1)
tuple2 = [random.randint(0, 50) for v in range(8)]
print(tuple2)
sum_tuple1 = sum(tuple1)
sum_tuple2 = sum(tuple2)
# print(sum_tuple1)
# print(sum_tuple2)
if sum_tuple1 >= sum_tuple2:
    print(
        f"la somme des éléments du tuple1 ({sum_tuple1}) est plus grand ou égal à la somme des elements du tuple2  ({sum_tuple2})")
else:
    print(
        f"la somme des éléments du tuple2 ({sum_tuple2}) est plus grand que la somme des elements du tuple1  ({sum_tuple1})")
```
