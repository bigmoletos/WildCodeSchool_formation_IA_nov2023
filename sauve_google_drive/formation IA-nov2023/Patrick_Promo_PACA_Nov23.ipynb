# Le métier de Data Analyst

### Le métier et le domaine

**" Sans les données vous êtes juste quelqu'un avec une opinion " Edward Deming**

**L'analyse des données** : Il y a beaucoup de définitions, mais en fin de compte, il s'agit du processus d'inspection, de nettoyage, de transformation et de modélisation des données pour la prise de décision stratégique.

**Data analyst** : Personne dont le travail consiste à examiner des données afin d'extraire des informaions utiles en de tirer des conclusions ou aider à la prise de décision, en exploitant le vécu et en prédisant (prévoir) le futur. Il s'agit d'un élément clé de toute équipe de data science. Le data analyst travaille avec des données sous toutes leurs formes et avec des outils variés. Pour le Data analyst la phrase "une image vaut mille mots" se transforme en "un graphique vaut mille lignes", il se sert donc de visualisations. Visualiser permet d'améliorer notre manière d'interpréter nos données. Il peut travailler de manière indépendante, ou en équipe avec des statisticiens, chercheurs... Les compétences qui lui seront demandées dépendront de l'organisation, en fonction des besoins parfois évolutifs (recherche de données, analyses, gouvernance, Data science, management de projet...). **Bref, un data analyst doit être curieux pour apprendre un maximum et continuer à apprendre.**

**Exemples d'applications**
- Optimisation du nombre de places dans les avions, hôtels, etc.
- Organisation des rayons dans les supermarchés
Organisation de campagne de publicité
- Diagnostic médical

**International Data Corporation (IDC) prévoit un besoin important de personnes dotées de compétences analytiques approfondies.**

On peut commencer à s'identifier en tant que data analyst si :
- Participe à la collecte, à l'inspection et au nettoyage de données en vu de la création de rapports (graphiques, diagrammes, rapports)
- Participer à la l'amélioration des processus sous un angle data driven
- Capacité à créer des modèles à partir des données

Les compétences clé d'un Data analyst :
- Statistiques/algèbre linéaire
- Tableur et outils BI
- SGBD(R ou NoSQL)
- Langage de programmation : Python, R
- Les plus
  - Machine learning
  - Plateforme Big Data : Hadoop

## Les tâches et compétences à avoir


**Identifier des données**

Un data analyst doit apprendre et savoir identifier des données pour valoriser toute donnée utile.

![chaise](https://cdn.shopify.com/s/files/1/0494/9599/8620/products/CLO2-BOIS-3-4.jpg?v=1609867331)

Exemple pour une chaise : Quels données voyez-vous ? couleur, matériau, type de meuble, dimension, poids, fournisseurs, localisation fournisseurs, fréquence d'approvisation, date de commande/d'expédition.

**Un data analyst doit savoir gérer les données manquantes en :**
- Les détectants
- Évaluant leurs impacts - significatif ou non => Est-ce que ça vaut le coup de les traiter ?
- Choisir une solution pour les traiter

Il doit s'intéresser aux valeurs possibles d'un type de donnée (string, date, number, booléen) et à la manière dont cette dernière est stockée.

L'organisation des données implique de différencier les données stockées de celles qui ne le sont pas, et de savoir quels calculs ou transformations (+,-,x,/, concaténation, fusion, conditions, conversion (dates, monnaies)) seront à effectuer pour utiliser les données (par exemple pour un rapport).

**Interpréter des données**

**Savoir interpréter les données est une qualité essentielle de tout bon data analyst.** En effet, souvent vous êtes face à une quantité importante de données et qui peuvent être très complexes.

![menu restaurant](https://img.freepik.com/vecteurs-libre/modele-menu-restaurant-rustique-plat-biologique-photo_52683-62703.jpg?w=2000)

**Un analyste de données travaille avec des règles métiers et des données pour établir des rapports, ce qui ne diffère en rien de la préparation d'une recette qui comprend des ingrédients et des instructions.** Lorsque vous êtes prêt à préparer un nouveau plat à partir d'une recette, la première chose que vous devez faire est acquérir l'ensemble des ingrédients dont vous avez besoin. Vous faites donc une liste des ingrédients que vous n'avez pas ou que vous avez déjà.

**Lorsque vous travaillez avec des données vous devez toujours créer une liste qui vous indique ce que vous avez, ce que vous devez trouver ou ce que vous devez créer.**

Une fois que vous avez déterminé la liste des données, il est maintenant temps de faire vos achats. Tout comme dans l'allée de l'épicerie, vous parcourez les données de vos différents systèmes ou feuilles de calcul. Vous pouvez trouver un million de choses dont vous avez besoin pendant vos achats, mais, souvenez-vous **du besoin premier pour réaliser votre recette et garder cela en tête quand vous cherchez des données**.


**Savoir utiliser les jointures**

Les jointures sont parfois la seule façon d'obtenir certains rapports. Il faut consacrer au maximum de temps aux jointures interne et externe.


**Savoir les graphiques de base*

Le Data Analyst doit savoir créer des graphiques de base.

**Savoir utiliser les tableaux croisés dynamiques**

Il est important de savoir créer et utiliser des tableaux croisés dynamiques.


## Quelques conseils et astuces

**Mener correctement un projet Data Analyse**

La démarche à suivre est la suivante :
- Préciser le besoin en données et l'objectif business
- Déterminer les règles métiers
- Chercher les données nécessaires pour atteindre l'objectif et répondre aux besoins
- Analyser les données

**Bonnes pratiques**

- Ne jamais deviner > prouver
- Définir le processus de vérification
- Organiser les réunions de travail
- Réaliser la documentation
- Processus d'amélioration continue

**Gouvernance de données**

Elle permet de garantir que les données disponibles sont de qualité, utilisables, sécurisées, intègres et respectent les règles métiers. Dans un certains nombres d'entreprise il peut ne pas y en avoir, il peut être demander au data analyst de régider le plan de gouvernance des données.

**Éviter les erreurs fréquentes**

[Erreurs](https://openclassrooms.com/fr/courses/7410486-nettoyez-et-analysez-votre-jeu-de-donnees/7450561-reperez-les-differents-types-derreurs)

**Flux de travail**

Tenir compte du flux de travail permet de mieux mener son analyse.

**Principes et méthodes de nettoyage de données**

Le but étant de supprimer les erreurs, de valider les données et les standardiser.

- Identifier les données : incomplètes, incorrectes, inexactes, non pertinentes, en doublon
- Effectuer un traitement : remplacer, modifier, supprimer

**Règle métiers**

Les règles métiers sont des éléments influençant dans la prise de décision, et différentes de la logique système. Pour bien l'appréhender il faut être en mesure de :
- Connaître les règles de saisie des données
- Le moment auquel elles doivent être entrée
- Le calcul à effectuer sur les données et son incidence sur les règles

**Créer un dictionnaire de données**

Un dictionnaire de données est un ensemble de données de référence nécessaire à la conception d'une base de données. Il permet de mapper les données en définissant les données et les relations qui les lient.

**Créer un fichier readme**

Il est utile parce qu'il permet de préciser des éléments importants pour effectuer le travail correctement. Il permet d'éviter aux utilisateurs de commettre des erreurs.

**Pour apprendre un langage de programmation, une méthode efficace est de lister les actions qu'on veut effectuer et de retrouver la liste des syntaxes de base permettant de le réaliser.** Il est important de faire des testes sur les syntaxes pour découvrir ce qui marche et ce qui ne marche pas, et pourquoi. Il vaudra évidemment s'habituer à faire des recherches sur le web ou dans des ouvrages de références.
