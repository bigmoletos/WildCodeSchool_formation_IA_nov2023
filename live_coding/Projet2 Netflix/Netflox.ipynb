```python
df = (
    df[
        [
            "Series_Title",
            "Released_Year",
            "Runtime",
            "Genre",
            "IMDB_Rating",
            "Director",
            "Star1",
            "Star2",
            "Star3",
            "Star4",
            "No_of_Votes",
            "Gross",
        ]
    ]
    .dropna()
    .reset_index(drop=True)
)
```

```python
df.head()
```

```python
df.info()
```

```python
df.Released_Year.astype("int")
```

```python
# transforme la date en int
listeReleased_Year = []
for element in df.Released_Year:
    try:
        valeur = int(element)
        listeReleased_Year.append(valeur)
    except:
        listeReleased_Year.append(np.nan)
```

```python
df.Runtime.str.split(" "):
```

```python
listeRuntime = []
for element in df.Runtime.str.split(" "):
    try:
        listeRuntime.append(element[0])
    except:
        listeRuntime.append(np.nan)
```

```python
df.Gross.str.replace(",", "").astype("int")
```

```python
df["Released_Year_clean"] = listeReleased_Year
df["Runtime_Clean"] = listeRuntime
df["Gross_clean"] = df.Gross.str.replace(",", "").astype("int")
```

```python
df
```

```python
dfFinal = (
    df.drop(["Released_Year", "Runtime", "Gross"], axis=1)
    .dropna()
    .reset_index(drop=True)
)
```

```python
dfFinal
```

```python
pd.get_dummies(dfFinal.Genre)
pd.get_dummies(dfFinal.Director)
pd.get_dummies(dfFinal.Star1)
pd.get_dummies(dfFinal.Star2)
pd.get_dummies(dfFinal.Star3)
pd.get_dummies(dfFinal.Star4)
```

```python
dfFeatures = pd.concat(
    [
        pd.get_dummies(dfFinal.Genre),        pd.get_dummies(dfFinal.Director),        pd.get_dummies(
            dfFinal.Star1),        pd.get_dummies(dfFinal.Star2),        pd.get_dummies(dfFinal.Star3),
        pd.get_dummies(dfFinal.Star4),        dfFinal["Series_Title"],
        dfFinal["IMDB_Rating"],        dfFinal["No_of_Votes"],        dfFinal["Released_Year_clean"],        dfFinal["Runtime_Clean"],
        dfFinal["Gross_clean"],
    ],

    axis=1,
)
```

```python
dfFeatures
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=15)
```

```python
neigh.fit(dfFeatures.drop("Series_Title", axis=1).values)
```

```python
# modele knn voisin les plus proches
distances, index = neigh.kneighbors(dfFeatures.drop("Series_Title", axis=1).values[0].reshape(1, -1)
)
```
