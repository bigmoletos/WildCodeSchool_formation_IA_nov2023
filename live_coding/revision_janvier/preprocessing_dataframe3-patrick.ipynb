- environ 68 % des valeurs qui vont tre tires vont tre comprises entre zro plus un cart-type et zro moins un cart-type
- 95 % environ des valeurs qui vont tre tires ou bien qui vont faire partie d'un chantillon issu d'une population qui suit une loi normale vont tre comprises entre zro plus deux carts-type et zro moins deux carts-type
- environ 99,7 % des valeurs d'un chantillon issus d'une population qui suit une loi normale vont tre comprises entre zro plus trois carts-type et zro moins trois carts-type

Si on tire d'une population qui suit une loi normale un chantillon reprsentatif, cet chantillon suivra une loi normale galement, avec les caractristiques associes. Par contre, si la population suit une autre loi, autre que la loi normale, alors, pour que le thorme central limite soit applicable, il faudrait que la taille des chantillons soit au minimum gale  30, Scikit learn prfrant une marge minimale  50 individus pour faire l'approximation d'une loi normale.

```python
df = pd.DataFrame({"valeur" : [1, 2, 3, 5, 3, 1, 3, 4, 6, 3, 8, 9, 10],
              "categorie" : ["A", "A", "B", "B", "B", "A", "B", "B", "C", "B", "C", "C", "C"]})
```

```python
df
```

```python
sns.boxplot(data = df, x = "valeur", y = "categorie")
```

```python
from statsmodels.formula.api import ols

traitement = ols("valeur ~ categorie", df)
```

```python
traitement
```

```python
from statsmodels.stats.anova import anova_lm

resultat = anova_lm(traitement.fit())
resultat
```

```python
if resultat.loc["categorie", "PR(>F)"] > 0.05:
  print("Il n'y a pas de corrlation entre la variable numrique et catgorielle")
elif resultat.loc["categorie", "PR(>F)"] < 0.05:
  print("Il a une corrlation entre la variable numrique et catgorielle")
```

```python
def anova(dataframe, colonneNumerique, colonneQualitative, p_value = 0.05):
  import pandas as pd
  from statsmodels.formula.api import ols
  from statsmodels.stats.anova import anova_lm

  traitement = ols(f"{colonneNumerique} ~ {colonneQualitative}", dataframe)
  traitement.fit()
  resultat = anova_lm(traitement.fit())
  print(resultat)
  print("La p-value vaut", resultat.loc[f"{colonneQualitative}", "PR(>F)"])

  if resultat.loc[f"{colonneQualitative}", "PR(>F)"] > p_value:
    print(f"Il n'y a pas de corrlation entre la colonne {colonneNumerique} et {colonneQualitative}")
  elif resultat.loc[f"{colonneQualitative}", "PR(>F)"] < p_value:
    print(f"Il a une corrlation entre la colonne {colonneNumerique} et {colonneQualitative}")
```
